```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

```python
#Distribution Plot
sns.distplot(crash_df['not_distracted'])
```

```python
#Joint Plot to compare two distribution

sns.jointplot(x='speeding', y = 'alcohol',data=crash_df,kind='reg')
#sns.jointplot(x='speeding', y = 'alcohol',data=crash_df,kind='kde')
```

```python
#KDE Plot
sns.kdeplot(crash_df['alcohol'])
```

```python
#Pair plot  - It is going to plot relationsship across numberical variables
#sns.pairplot(crash_df)

tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df,hue='sex',palette='Blues')
```

```python
#Styling
#sns.set_style('white')
sns.set_style('darkgrid')
plt.figure(figsize=(8,4))
sns.set_context('paper',font_scale=1.4)
#sns.set_context('talk',font_scale=1.4)

sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

Categorical Plot


```python
#Bar plot

sns.barplot(x='sex',y='total_bill',data=tips_df)
```

```python
#Count Plot
sns.countplot(x='sex',data=tips_df)
```

```python
#Box plot

sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

```python
 sns.swarmplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

```python
#Palettes   To change colours
```

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex',palette='magma')
```

```python
#Heat Maps

plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)

crash_matrix = crash_df.corr()
sns.heatmap(crash_matrix,annot=True)
```

```python
#Regression plot

plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips_df)
```
